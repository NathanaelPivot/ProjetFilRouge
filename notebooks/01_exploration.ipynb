Objectif de ce notebook : charger le dataset Steam Games et faire un premier
diagnostic rapide (taille, colonnes, types, valeurs manquantes, doublons) avant

Avant d'exécuter ce notebook : téléchargez `games.csv` depuis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
```

## Chargement des données

```python
DATA_PATH = '../data/raw/games.csv'

df = pd.read_csv(DATA_PATH)
print(f"Dimensions du dataset : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

## Valeurs manquantes

Nombre et pourcentage de valeurs manquantes par colonne, triées par importance.

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_summary = pd.DataFrame({
    'nb_manquantes': missing,
    'pct_manquantes': missing_pct
}).sort_values('pct_manquantes', ascending=False)

missing_summary[missing_summary['nb_manquantes'] > 0]
```

## Doublons

```python
nb_doublons = df.duplicated().sum()
print(f"Nombre de lignes dupliquées : {nb_doublons}")

# Vérifie aussi les doublons sur l'identifiant unique si la colonne existe
if 'AppID' in df.columns:
    print(f"Doublons sur AppID : {df['AppID'].duplicated().sum()}")
```
